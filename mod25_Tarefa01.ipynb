1. Marque quais desses métodos/algoritmos muito populares em ciência de dados são baseados no uso de derivada:

    1. Método Mínimos Quadrados
    2. Gradiente descendente
    3. Newton Raphson
    4. CART (Árvore de decisão)

___________

Método dos Mínimos Quadrados: ✅ utiliza derivadas para minimizar a soma dos erros quadrados. A derivada da função de erro é usada para encontrar os coeficientes que minimizam essa função.

Gradiente Descendente: ✅ esse método é diretamente baseado em derivadas. Ele calcula a derivada da função de custo para ajustar os parâmetros na direção que minimiza a função de erro.

Newton-Raphson: ✅ método numérico que utiliza tanto a primeira quanto a segunda derivada para encontrar as raízes de funções, incluindo o uso em otimização para minimizar funções de custo.

CART (Árvore de Decisão): ❌ o algoritmo de árvores de decisão não utiliza derivadas. Ele se baseia em critérios como entropia, ganho de informação ou índice Gini para dividir os nós, sem precisar calcular derivadas.

2. Dada uma base de dados com uma variável resposta $y$ e um conjunto de variáveis explicativas. Considere uma estrutura de um modelo de regressão. Explique com suas palavras por que não é possível obter parâmetros que forneçam um erro quadrático médio (EQM) menor que o obtido com estimadores de mínimos quadrados.

A razão pela qual não é possível obter parâmetros que forneçam um Erro Quadrático Médio (EQM) menor do que o obtido com os estimadores de mínimos quadrados se deve ao fato de que o Método dos Mínimos Quadrados foi projetado exatamente para minimizar essa métrica.

O OLS minimiza o EQM por definição, e qualquer tentativa de ajuste diferente (sem alterar o modelo ou as suposições) só pode igualar ou piorar o EQM, mas nunca melhorá-lo.

**Demonstrando em Código**:

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

# 1. Gerando dados simulados (linear com ruído)
np.random.seed(42)
n_samples = 100
X = 2 * np.random.rand(n_samples, 1)  # Variável explicativa
y = 4 + 3 * X + np.random.randn(n_samples, 1)  # Variável resposta (com ruído)

# 2. Ajustar o modelo de regressão linear com OLS (Mínimos Quadrados)
model = LinearRegression()
model.fit(X, y)

# Prever os valores com o modelo ajustado
y_pred = model.predict(X)

# 3. Calcular o Erro Quadrático Médio (EQM)
eqm = mean_squared_error(y, y_pred)

# Plotando os dados e a reta ajustada
plt.figure(figsize=(8, 6))
plt.scatter(X, y, label="Dados reais", color='blue')
plt.plot(X, y_pred, label="Reta ajustada (OLS)", color='red')
plt.title(f"Regressão Linear com OLS\nEQM = {eqm:.4f}")
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

eqm
```

**Análise**:

O Método dos Mínimos Quadrados Ordinários (OLS), embora eficiente para encontrar os parâmetros que minimizam o Erro Quadrático Médio (EQM), não possui mecanismos para controlar ou reduzir o overfitting.
